```python
import json
response = endpoints.teamgamelogs.TeamGameLogs(season_nullable='2021-22')
content = json.loads(response.get_json())
results = content['resultSets'][0]
headers = results['headers']
rows = results['rowSet']
record = pd.DataFrame(rows, columns = headers)
record.head()
```

```python
record = record.groupby(['TEAM_ABBREVIATION'])['WL'].value_counts()
record = pd.DataFrame(record).rename(columns = {'WL': 'W'})
record.head()
```

```python
record = pd.pivot_table(record, values = 'W', index = 'TEAM_ABBREVIATION', columns = 'WL')
```

```python
team_eff = team_eff.merge(record, left_on = 'TEAM', right_on = 'TEAM_ABBREVIATION')
team_eff.head() 
```
